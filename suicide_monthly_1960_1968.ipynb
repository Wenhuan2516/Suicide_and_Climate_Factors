```python
import io

# Replace 'file.csv' with the path to your CSV file
file_path = 'suicide_death/mort' + year + '.csv'

# Open the file with the desired encoding and error handling
with open(file_path, 'r', encoding='utf-8', errors='replace') as file:
    # Read the entire file into a string
    file_content = file.read()

# Create a StringIO object from the file content
file_io = io.StringIO(file_content)

# Read the CSV file using pandas
df = pd.read_csv(file_io)

# Now you can work with the DataFrame 'df'
```

```python
valid_columns = ['datayear', 'countyrs','staters', 'stateoc','countyoc', 'monthdth','sex','race', 'ager12','ucod', 'ucr281', 'ucr69', 'ucr34']
```

```python
df_valid = df[valid_columns]
df_valid.head()
```

### Cleaning the columns

```python
df_valid = df_valid.rename(columns = {'datayear': 'year', 'countyrs': 'county_residence', 'staters': 'state_residence',
                                     'stateoc':'state_death', 'countyoc': 'county_death', 'ager12': 'age', 'ucod': 'death_cause',
                                     'ucr281': 'death_cause_281', 'ucr69': 'death_cause_69', 'ucr34': 'death_cause_34'})
```

```python
df_valid.head()
```

```python
 df_valid['year'] = year
```

```python
df_valid['county_residence'] = df_valid['county_residence'].astype(str)
df_valid['state_residence'] = df_valid['state_residence'].astype(str)
df_valid['county_death'] = df_valid['county_death'].astype(str)
df_valid['state_death'] = df_valid['state_death'].astype(str)
```

```python
def formatCountyCode(code):
    return code.rjust(5, '0')
```

```python
def formatStateCode(code):
    return code.rjust(2, '0')
```

```python
df_valid['county_residence'] = df_valid['county_residence'].apply(formatCountyCode)
df_valid['state_residence'] = df_valid['state_residence'].apply(formatStateCode)
df_valid['county_death'] = df_valid['county_death'].apply(formatCountyCode)
df_valid['state_death'] = df_valid['state_death'].apply(formatStateCode)
```

```python
replace_month = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 
                10: 'Oct', 11: 'Nov', 12: 'Dec'}
df_valid['monthdth'] = df_valid['monthdth'].replace(replace_month)
df_valid.head()
```

```python
replace_sex = {1: 'Male', 2: 'Female'}
df_valid['sex'] = df_valid['sex'].replace(replace_sex)
```

#### replace race information

```python
replace_race = {0: 'other', 1: 'White', 2: 'Black', 3: 'other', 4: 'other', 5: 'other', 6: 'other', 7: 'other', 8: 'other'}
df_valid['race'] = df_valid['race'].replace(replace_race)
df_valid.head()
```

### count the suicide death in each month/county/race/sex

```python
def ageRange(age):
    if age <= 4:
        return 'age 0-24'
    elif age >= 5 and age <= 8:
        return 'age 25-64'
    elif age >= 9 and age <= 11:
        return '65 and older'
    else:
        return np.nan
```

```python
df_suicide['age'].dtype
```

```python
df_suicide['age_range'] = df_suicide['age'].apply(ageRange)
df_suicide.head()
```

##### From my research, I found code '32900' in death_cause_281 represent firearm suicide death, and the rest of code represetn non-firearm death. So we can divide df_suicide to be 2 seprate dataframe: one is firearm and the other one is non-firearm cases. 

```python
df_suicide.columns
```

```python
column_list = ['year', 'county_residence', 'state_residence', 'state_death','county_death', 'monthdth', 
               'sex', 'race', 'age', 'age_range', 'death_cause_281']
```

```python
df_suicide = df_suicide[column_list]
```

```python
df_firearm = df_suicide[df_suicide['death_cause_281'] == 32900]
df_nonfirearm = df_suicide[df_suicide['death_cause_281'] != 32900]
```

```python
df_firearm['death_cause_281'] = 'firearm_suicide'
df_nonfirearm['death_cause_281'] = 'nonfirearm_suicide'
```

```python
df_firearm.shape
```

```python
df_nonfirearm.shape
```

```python
df_firearm = df_firearm.rename(columns = {'monthdth': 'month'})
df_nonfirearm = df_nonfirearm.rename(columns = {'monthdth': 'month'})
```

```python
df_suicide = df_suicide.rename(columns = {'monthdth': 'month'})
```

```python
def convertCode(code):
    if code == 32900:
        return 'firearm_suicide'
    else:
        return 'nonfirearm_suicide'
```

```python
df_suicide['death_cause_281'] = df_suicide['death_cause_281'].apply(convertCode)
df_suicide.head()
```
